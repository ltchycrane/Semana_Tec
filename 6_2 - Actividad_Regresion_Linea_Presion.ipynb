#**Presion Arterial**

1. Carga los datos del archivo **presion.csv**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import median_absolute_error
```

```python
df = pd.read_csv('data/presion.csv')
df.head()
```

2. Gráfica la información en el eje de las x la edad  y en el eje y la presión ( alta y baja) 


```python
# group by age
df_age = df.groupby('edad').mean()

# age vs Average of ap_hi
age_vs_ap_hi = df_age['Average of ap_hi']
age_vs_ap_hi.plot(figsize=(10, 5), title='Age vs Average of ap_hi', xlabel='Age', ylabel='Average of ap_hi')
```

```python
# age vs Average of ap_lo
age_vs_ap_lo = df_age['Average of ap_lo']
age_vs_ap_lo.plot(figsize=(10, 5), title='Age vs Average of ap_lo', xlabel='Age', ylabel='Average of ap_lo')
```

3. Genera una regresión líneal para obtener una aproximación de la ecuación 
$$
y = ax + b
$$
donde $ a $ se conoce comúnmente como **pendiente**, y $ b $ se conoce comúnmente como **intersección**.

¿Cuál es el valor de a y cuál es el valor de b para la presión alta y baja?

```python
# Linear Regression between age and ap_hi
X = df_age.index.values.reshape(-1, 1)
y = df_age['Average of ap_hi'].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor_hi = LinearRegression()
regressor_hi.fit(X_train, y_train)
y_pred = regressor_hi.predict(X_test)
df_age['ap_hi_pred'] = regressor_hi.predict(X)
```

```python
# Linear Regression between age and ap_lo
X = df_age.index.values.reshape(-1, 1)
y = df_age['Average of ap_lo'].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor_lo = LinearRegression()
regressor_lo.fit(X_train, y_train)
y_pred = regressor_lo.predict(X_test)
df_age['ap_lo_pred'] = regressor_lo.predict(X)
```

Genera una función que calcule lo anterior

```python
edad_calcular=76

def presionLo(edad):
    # Use the regressor_lo model to predict the ap_lo
    return regressor_lo.predict([[edad]])[0][0]

presionLo(edad_calcular)
```
