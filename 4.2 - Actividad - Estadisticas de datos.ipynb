# Insurance dataset

El dataset contiene información demográfica sobre los asegurado en una compañía de seguros:

- age: edad del asegurado principal
- sex: género del asegurado. female o male
- bmi: índice de masa corporal
- children: numero de niños que estan cubiertos con la poliza. 
- smoke: si fuma el beneficiario. yes/no
- region: dónde vive el beneficiario. Estos datos son de Estados Unidos. Regiones disponibles: northeast, southeast, southwest, northwest
- charges: costo del seguro.


# Actividad
Carga el dataset  **`data/insurance.csv`** y haz un análisis estadístico de las variables.

```python
# Read the data
import pandas as pd
import numpy as np
path = "./data/"
df = pd.read_csv(path + "insurance.csv")
```

```python
# Print the first 5 rows of the dataframe.
df.head()
```

```python
# Statistics of the numerical columns
df.describe()
```

```python
# ¿Cómo se correlacionan las variables numéricas entre sí?
df.corr()
# ¿Cuáles son las variables que tienen más correlación con los costos del seguro?
# De acuerdo a la correlacion, las variables age y charges
```

```python
# ¿Cuántas asegurados son hombres y cuántas son mujeres?
print("Hombre:",len(df[df["sex"]=="male"]))
print("Mujeres:",len(df[df["sex"]=="female"]))
print("Total:",len(df["sex"]))
```

```python
# En promedio, ¿quién paga más de cuota de seguro, los fumadores o no fumadores? Muéstralo con los datos

# Get the difference of the averagre of the column charges bewteen smokers and non smokers
print("Diferencia de promedio de cuota de seguro entre fumadores y no fumadores:",df[df["smoker"]=="yes"]["charges"].mean()-df[df["smoker"]=="no"]["charges"].mean())
```

```python
# ¿Cuál es la región con más variabilidad en la cuota de seguro? Muéstralo con los datos

# Get the standard deviation of the column charges by region
print("Región con mayor desviación estandar:",df.groupby("region")["charges"].std().idxmax())
```
